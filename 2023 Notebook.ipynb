id is the primary key column for this table.
Each row of this table indicates the ID of an employee, their name, compensation, and the ID of their manager.
 

Write an SQL query to find the employees who earn more than their managers.

Return the result table in any order.

```python
%sql

--INSERT YOUR SQL COMMAND HERE

SELECT id, employee_name, compensation, manager_id 
FROM EMPLOYEE A, EMPLOYEE B
WHERE 
A.id = B.manager_id
AND B.compensation > A.compensation
```
